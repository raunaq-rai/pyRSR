```python
from astropy.io import fits
import numpy as np
from PyRSR.plotting import plot_spectrum_with_2d

specfile = "../../data/excels-uds04-v4_g395m-f290lp_3543_63107.spec.fits"

with fits.open(specfile) as hdul:
    hdul.info()
```

```python
specfile = "../../data/excels-uds04-v4_g395m-f290lp_3543_63107.spec.fits"
with fits.open(specfile) as hdul:

    # 1D spectrum
    t = hdul["SPEC1D"].data
    lam = np.asarray(t["wave"])     # [m]
    fnu = np.asarray(t["flux"])     # [Jy]
    enu = np.asarray(t["err"])      # [Jy]

    # 2D spectrum (EXCELS format)
    sci2d = hdul["SCI"].data      # shape (ny, nx)

lam_min = 3.0
lam_max = 5.0

mask = (lam >= lam_min) & (lam <= lam_max)

lam_crop = lam[mask]
fnu_crop = fnu[mask]
enu_crop = enu[mask]

sci2d_crop = sci2d[:, mask]

# ---------- Plot ----------
fig = plot_spectrum_with_2d(
    lam_um   = lam_crop,
    flux_uJy = fnu_crop,
    err_uJy  = enu_crop,
    sci2d    = sci2d_crop,
    title="2D + 1D Spectrum (35.5 m)",
    xlim=(lam_min, lam_max),
)
```
